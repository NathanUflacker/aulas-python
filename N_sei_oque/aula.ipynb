# Estrutura com while

```python
data = {
    'primeiro_nome': ['joão', 'ana', 'laura'],
    'idade': [16, 29, 20]
}

df = pd.DataFrame(data)
df
```

```python
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        print(row['primeiro_nome'], row['idade'])

    except StopIteration:
        break
```

```python
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        if row['Tipo da Falha'] == 'No Failure':
            print(f"UDI: {row['UDI']}")
            print(f"ID Produto: {row['ID Produto']}")

    except StopIteration:
        break
```

```python
contagem_falhas = {}
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        tipo_falha = row['Tipo da Falha']

        if tipo_falha in contagem_falhas:
            contagem_falhas[tipo_falha] += 1
        else:
            contagem_falhas[tipo_falha] = 1

    except StopIteration:
        break

    
for tipo_falha, contagem in contagem_falhas.items():
    print('Tipo da Falha: ', tipo_falha, "Quantidade: ", contagem)
```

```python
# Calculando a média usando for e while

soma_desgaste = 0
count_m = 0

for index, row in df.iterrows():
    if row['Tipo'] == 'M':
        soma_desgaste += row['Desgaste Ferramenta [min]']
        count_m += 1

if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print("A Média do desgaste da ferramenta para o tipo M é: ", media_desgaste)
else:
    print('Não há dados')
```

```python
soma_desgaste = 0
count_m = 0
iterator = df.iterrows()

while True:
    try:
        index, row = next(iterator)
        if row['Tipo'] == 'M':
            soma_desgaste += row['Desgaste Ferramenta [min]']
            count_m += 1

    except StopIteration:
        break

if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print("A Média do desgaste da ferramenta para o tipo M é: ", media_desgaste)
else:
    print('Não há dados')
```
